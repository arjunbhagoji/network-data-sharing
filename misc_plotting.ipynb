```python
import numpy as np
import argparse

import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
import configparser
```

```python
config = configparser.ConfigParser()
config.read_file(open('configs/default.cfg'))
```

```python
def load_and_plot(config,day_name,attack_type,classifier,maintain_ratio,seq_select):
    if config['DATA']['input_type']=='CIC-IDS-2017':
        output_dir_name=config['DATA']['output_dir']+'/'+config['DATA']['input_type']+'/'+day_name+'-'+attack_type
    out_file_name=classifier
    if maintain_ratio:
        out_file_name+='_mr'
    if seq_select:
        out_file_name+='_seq'
    out_file_name=output_dir_name+'/'+out_file_name+'.txt'
    a=np.loadtxt(out_file_name,delimiter=',')
    
    fig, ax = plt.subplots()
    ax.errorbar(a[:,0], a[:,1],a[:,2],color='green', marker='o', linestyle='dashed',linewidth=2, markersize=5,label='Acc.')

    ax.set(xlabel='Num. splits', ylabel='Classification Accuracy',
           title='Accuracy variation ({})'.format(classifier))

    ax.legend()
#     plt.savefig('outputs/images/.pdf' % ())
    plt.show()
```

```python
day_name='Friday'
attack_type='DDoS'
classifier='SVM'
maintain_ratio=True
seq_select=False
```

```python
load_and_plot(config,day_name,attack_type,classifier,maintain_ratio,seq_select)
```
